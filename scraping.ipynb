```python
import os
import csv
import requests
import xml.etree.ElementTree as ET
from collections import defaultdict

# ----------------------------------------------------------------------
# Parameters
BASE_URL = "https://www.martinbrake.de/timelines/data/bahn/"
OUTPUT_DIR = "bahn_kml"
YEAR = 1885  # change this if you want a different year snapshot

# Raw types and states from /timelines/js/bahn.js
RAW_TYPES = [
    {"id": "NE",  "gl": [1, 2]},  # Nebenstr. electrified / not electrified
    {"id": "ND",  "gl": [1, 2]},  # Nebenstr.
    {"id": "HE",  "gl": [1, 2]},  # Hauptstr. electrified
    {"id": "HD",  "gl": [1, 2]},  # Hauptstr.
    {"id": "K",   "gl": [1]},     # Klein-/Schmalspurbahn
    {"id": "HGV", "gl": [2]},     # High-speed railway (≥200 km/h)
]

COUNTRIES = [
    {"id": "NRW"},  # Nordrhein‑Westfalen
    {"id": "HE"},   # Hessen
    {"id": "NI"},   # Niedersachsen
    {"id": "SH"},   # Schleswig‑Holstein
    {"id": "RPL"},  # Rheinland‑Pfalz+Saarland
    {"id": "BW"},   # Baden‑Württemberg
    {"id": "BY"},   # Bayern
    {"id": "TH"},   # Thüringen
    {"id": "SNA"},  # Sachsen‑Anhalt
    {"id": "SN"},   # Sachsen
    {"id": "BR"},   # Brandenburg/Berlin
    {"id": "MP"},   # Mecklenburg‑Vorpommern
]

# ----------------------------------------------------------------------
def generate_file_ids():
    """
    Generate all KML file identifiers from states and raw_types.
    """
    ids = []
    for country in COUNTRIES:
        for rtype in RAW_TYPES:
            for gl in rtype["gl"]:
                ids.append(f"{country['id']}_{rtype['id']}{gl}")
    return ids

def download_kml(file_id, out_dir):
    """
    Download a KML file with headers to satisfy the host.
    """
    os.makedirs(out_dir, exist_ok=True)
    url = f"{BASE_URL}{file_id}.kml"
    headers = {
        "Referer": "https://www.martinbrake.de/timelines/bahn",
        "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) "
                      "AppleWebKit/537.36 (KHTML, like Gecko) "
                      "Chrome/120.0 Safari/537.36",
    }
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    path = os.path.join(out_dir, f"{file_id}.kml")
    with open(path, "wb") as f:
        f.write(response.content)
    return path

def parse_kml_for_year(kml_path, year):
    """
    Parse a KML file and return two objects:
      - a set of station coordinates (lon, lat)
      - a list of lines (each with start/end coords and the full polyline)
    Only lines that were open during the specified year are retained.
    """
    station_coords = set()
    lines = []
    root = ET.parse(kml_path).getroot()

    # iterate through all Placemark elements
    for placemark in root.findall(".//Placemark"):
        # check TimeSpan for validity in 'year'
        timespan = placemark.find(".//TimeSpan")
        if timespan is None:
            continue
        begin = timespan.find("begin").text if timespan.find("begin") is not None else None
        end   = timespan.find("end").text if timespan.find("end") is not None else None

        def to_year(s):
            if not s:
                return None
            try:
                return int(s.split("-")[0])
            except Exception:
                return None

        begin_year = to_year(begin)
        end_year   = to_year(end)

        if begin_year is None:
            continue
        if begin_year > year:
            continue
        if end_year is not None and end_year < year:
            continue

        # extract coordinates of the LineString
        coords_el = placemark.find(".//LineString/coordinates")
        if coords_el is None or not coords_el.text:
            continue
        coord_pairs = []
        for c in coords_el.text.strip().split():
            parts = c.split(",")
            if len(parts) >= 2:
                lon = float(parts[0])
                lat = float(parts[1])
                coord_pairs.append((lon, lat))
        if not coord_pairs:
            continue

        start_pt = coord_pairs[0]
        end_pt   = coord_pairs[-1]
        station_coords.add(start_pt)
        station_coords.add(end_pt)
        lines.append({"start": start_pt, "end": end_pt, "coords": coord_pairs})

    return station_coords, lines

def main():
    # Step 1: construct file IDs
    file_ids = generate_file_ids()

    # Step 2: download all KML files
    downloaded_paths = []
    for fid in file_ids:
        path = os.path.join(OUTPUT_DIR, f"{fid}.kml")
        if not os.path.exists(path):
            try:
                print(f"Downloading {fid} …")
                path = download_kml(fid, OUTPUT_DIR)
            except Exception as ex:
                print(f"Warning: could not download {fid}: {ex}")
                continue
        downloaded_paths.append(path)

    # Step 3: parse each KML and collect lines/stations for the given year
    all_station_coords = set()
    all_lines = []
    for kml_path in downloaded_paths:
        stations, lines = parse_kml_for_year(kml_path, YEAR)
        all_station_coords.update(stations)
        all_lines.extend(lines)
    print(f"Found {len(all_station_coords)} unique stations and {len(all_lines)} line segments for {YEAR}.")

    # Step 4: assign station IDs and export to CSV
    stations_list = list(all_station_coords)
    station_index = {coord: i for i, coord in enumerate(stations_list)}

    with open("stations_1885.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["station_id", "lon", "lat"])
        for sid, (lon, lat) in enumerate(stations_list):
            writer.writerow([sid, lon, lat])

    with open("railway_lines_1885.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["line_id", "start_station", "end_station", "coords"])
        for lid, line in enumerate(all_lines):
            start_sid = station_index[line["start"]]
            end_sid   = station_index[line["end"]]
            # Represent the entire polyline as a pipe‑separated string of lon:lat pairs
            coords_str = "|".join(f"{lon}:{lat}" for lon, lat in line["coords"])
            writer.writerow([lid, start_sid, end_sid, coords_str])

    print("CSV exports complete: stations_1885.csv and railway_lines_1885.csv")

if __name__ == "__main__":
    main()
```

```python
import os
import csv
import time
import math
import requests
import xml.etree.ElementTree as ET
from collections import defaultdict
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd

# ----------------------------------------------------------------------
# Parameters
BASE_URL = "https://www.martinbrake.de/timelines/data/bahn/"
OUTPUT_DIR = "bahn_kml"
YEAR = 1885
MAX_GEOCODES = 500  # Limit geocoding to first N stations

g = 6371.0  # Radius of the Earth in km

def haversine(p1, p2):
    from math import radians, sin, cos, sqrt, atan2
    lon1, lat1 = map(radians, p1)
    lon2, lat2 = map(radians, p2)
    dlon = lon2 - lon1
    dlat = lat2 - lat1
    a = sin(dlat / 2)**2 + cos(lat1) * cos(lat2) * sin(dlon / 2)**2
    c = 2 * atan2(sqrt(a), sqrt(1 - a))
    return g * c

RAW_TYPES = [
    {"id": "NE",  "gl": [1, 2]},
    {"id": "ND",  "gl": [1, 2]},
    {"id": "HE",  "gl": [1, 2]},
    {"id": "HD",  "gl": [1, 2]},
    {"id": "K",   "gl": [1]},
    {"id": "HGV", "gl": [2]},
]

COUNTRIES = [
    {"id": "NRW"}, {"id": "HE"}, {"id": "NI"}, {"id": "SH"}, {"id": "RPL"},
    {"id": "BW"}, {"id": "BY"}, {"id": "TH"}, {"id": "SNA"}, {"id": "SN"},
    {"id": "BR"}, {"id": "MP"},
]

def generate_file_ids():
    ids = []
    for country in COUNTRIES:
        for rtype in RAW_TYPES:
            for gl in rtype["gl"]:
                ids.append(f"{country['id']}_{rtype['id']}{gl}")
    return ids

def download_kml(file_id, out_dir):
    os.makedirs(out_dir, exist_ok=True)
    url = f"{BASE_URL}{file_id}.kml"
    headers = {
        "Referer": "https://www.martinbrake.de/timelines/bahn",
        "User-Agent": "Mozilla/5.0"
    }
    response = requests.get(url, headers=headers)
    response.raise_for_status()
    path = os.path.join(out_dir, f"{file_id}.kml")
    with open(path, "wb") as f:
        f.write(response.content)
    return path

def parse_kml_for_year(kml_path, year):
    station_earliest = dict()
    lines = []
    root = ET.parse(kml_path).getroot()

    for placemark in root.findall(".//Placemark"):
        timespan = placemark.find(".//TimeSpan")
        if timespan is None:
            continue
        begin = timespan.find("begin").text if timespan.find("begin") is not None else None
        end = timespan.find("end").text if timespan.find("end") is not None else None

        def to_year(s):
            if not s:
                return None
            try:
                return int(s.split("-")[0])
            except Exception:
                return None

        begin_year = to_year(begin)
        end_year = to_year(end)

        if begin_year is None or begin_year > year:
            continue
        if end_year is not None and end_year < year:
            continue

        coords_el = placemark.find(".//LineString/coordinates")
        if coords_el is None or not coords_el.text:
            continue

        coord_pairs = []
        for c in coords_el.text.strip().split():
            parts = c.split(",")
            if len(parts) >= 2:
                lon = float(parts[0])
                lat = float(parts[1])
                coord_pairs.append((lon, lat))

        if not coord_pairs:
            continue

        start_pt = coord_pairs[0]
        end_pt = coord_pairs[-1]
        dist = sum(haversine(coord_pairs[i], coord_pairs[i+1]) for i in range(len(coord_pairs)-1))

        for pt in (start_pt, end_pt):
            if pt not in station_earliest or station_earliest[pt] > begin_year:
                station_earliest[pt] = begin_year

        lines.append({"start": start_pt, "end": end_pt, "year_built": begin_year, "length_km": dist, "coords": coord_pairs})

    return station_earliest, lines

def reverse_geocode_all(coords):
    geolocator = Nominatim(user_agent="railway1885", timeout=10)
    reverse = RateLimiter(
        geolocator.reverse,
        min_delay_seconds=1,
        max_retries=3,
        error_wait_seconds=5
    )
    results = {}
    for i, coord in enumerate(coords):
        if i >= MAX_GEOCODES:
            results[coord] = ""
            continue
        try:
            location = reverse(coord, exactly_one=True, language="en")
            if location and hasattr(location, "raw"):
                city = location.raw.get("address", {}).get("town") or \
                       location.raw.get("address", {}).get("village") or \
                       location.raw.get("address", {}).get("city") or ""
                results[coord] = city
            else:
                results[coord] = ""
        except Exception as e:
            print(f"⚠️ Geocoding failed for {coord}: {e}")
            results[coord] = ""
    return results

def main():
    file_ids = generate_file_ids()
    downloaded_paths = []
    for fid in file_ids:
        path = os.path.join(OUTPUT_DIR, f"{fid}.kml")
        if not os.path.exists(path):
            try:
                print(f"Downloading {fid} …")
                path = download_kml(fid, OUTPUT_DIR)
            except Exception as ex:
                print(f"Warning: could not download {fid}: {ex}")
                continue
        downloaded_paths.append(path)

    station_years = dict()
    all_lines = []
    for kml_path in downloaded_paths:
        stations, lines = parse_kml_for_year(kml_path, YEAR)
        station_years.update(stations)
        all_lines.extend(lines)

    print(f"Found {len(station_years)} stations and {len(all_lines)} lines in {YEAR}")

    city_lookup = reverse_geocode_all(station_years.keys())

    # Write stations CSV and XLSX
    station_rows = []
    for sid, (lon, lat) in enumerate(station_years.keys()):
        station_rows.append([sid, lon, lat, station_years[(lon, lat)], city_lookup.get((lon, lat), "")])

    with open("stations_1885_new.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["station_id", "lon", "lat", "year_built", "city_name"])
        writer.writerows(station_rows)

    pd.DataFrame(station_rows, columns=["station_id", "lon", "lat", "year_built", "city_name"]).to_excel("stations_1885_new.xlsx", index=False)

    # Map coords to station IDs
    station_index = {coord: sid for sid, coord in enumerate(station_years.keys())}

    line_rows = []
    for lid, line in enumerate(all_lines):
        coords_str = "|".join(f"{lon}:{lat}" for lon, lat in line["coords"])
        line_rows.append([
            lid,
            station_index[line["start"]],
            station_index[line["end"]],
            line["year_built"],
            round(line["length_km"], 3),
            coords_str
        ])

    with open("railway_lines_1885_new.csv", "w", newline="", encoding="utf-8") as f:
        writer = csv.writer(f)
        writer.writerow(["line_id", "start_station", "end_station", "year_built", "length_km", "coords"])
        writer.writerows(line_rows)

    pd.DataFrame(line_rows, columns=["line_id", "start_station", "end_station", "year_built", "length_km", "coords"]).to_excel("railway_lines_1885_new.xlsx", index=False)

    print("✓ CSV and XLSX files written: stations_1885_new.*, railway_lines_1885_new.*")

if __name__ == "__main__":
    main()
```

```python
"""
Grid-based density of railway coordinates (point counts per square cell)

Input CSV: must have a column named 'coords' with values like:
  "lon:lat|lon:lat|lon:lat|..."

Output:
  1) grid_counts.csv  (cell center + count per cell)
  2) heatmap.png      (optional heatmap)

Run:
  python grid_density.py
(or just paste into a notebook / VSCode and run)

Adjust:
  CELL_SIZE_DEG to change grid resolution (0.1 ~ 10–11 km in latitude).
"""

import os
import math
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt


# ---------------------------
# USER SETTINGS
# ---------------------------
INPUT_CSV = r"C:\Users\Caleb\Downloads\scraping\railway_lines_1885_new.csv"

OUT_CSV = os.path.join(os.path.dirname(INPUT_CSV), "grid_counts.csv")
OUT_PNG = os.path.join(os.path.dirname(INPUT_CSV), "heatmap.png")

CELL_SIZE_DEG = 0.10   # grid cell size in degrees (lat/lon). Try 0.05 for finer.
MAKE_PLOT = True


# ---------------------------
# HELPERS
# ---------------------------
def parse_coords_string(s: str):
    """
    Parse one coords string of the form:
      "lon:lat|lon:lat|..."
    Returns list of (lon, lat) floats.
    """
    if s is None or (isinstance(s, float) and math.isnan(s)):
        return []

    s = str(s).strip()
    if not s:
        return []

    points = []
    for item in s.split("|"):
        item = item.strip()
        if not item:
            continue
        parts = item.split(":")
        if len(parts) != 2:
            continue
        try:
            lon = float(parts[0])
            lat = float(parts[1])
            points.append((lon, lat))
        except ValueError:
            continue
    return points


def build_grid_counts(df: pd.DataFrame, cell_size: float):
    """
    Returns:
      counts_df: rows for each populated cell with (row, col, lon_center, lat_center, count)
      grid_meta: dict with bounding box + grid dimensions
      grid_matrix: 2D numpy array (rows x cols) with counts (for plotting)
    """
    if "coords" not in df.columns:
        raise ValueError("CSV must contain a 'coords' column with 'lon:lat|lon:lat|...' values.")

    # Collect all points
    lons = []
    lats = []
    for s in df["coords"].values:
        pts = parse_coords_string(s)
        for lon, lat in pts:
            lons.append(lon)
            lats.append(lat)

    if not lons:
        raise ValueError("No coordinate points found in 'coords' column.")

    lons = np.array(lons, dtype=float)
    lats = np.array(lats, dtype=float)

    min_lon, max_lon = float(np.min(lons)), float(np.max(lons))
    min_lat, max_lat = float(np.min(lats)), float(np.max(lats))

    # Grid dimensions
    ncols = int(math.floor((max_lon - min_lon) / cell_size)) + 1
    nrows = int(math.floor((max_lat - min_lat) / cell_size)) + 1

    # Assign each point to a cell
    cols = np.floor((lons - min_lon) / cell_size).astype(int)
    rows = np.floor((lats - min_lat) / cell_size).astype(int)

    # Keep only points that land inside grid bounds (safety)
    m = (cols >= 0) & (cols < ncols) & (rows >= 0) & (rows < nrows)
    rows = rows[m]
    cols = cols[m]

    # Count points per (row, col)
    rc = pd.DataFrame({"row": rows, "col": cols})
    counts = rc.value_counts().reset_index(name="count")  # columns: row, col, count

    # Compute cell centers
    counts["lon_center"] = min_lon + (counts["col"] + 0.5) * cell_size
    counts["lat_center"] = min_lat + (counts["row"] + 0.5) * cell_size

    # Build dense matrix for plotting
    grid = np.zeros((nrows, ncols), dtype=int)
    grid[counts["row"].to_numpy(), counts["col"].to_numpy()] = counts["count"].to_numpy()

    grid_meta = {
        "min_lon": min_lon,
        "max_lon": max_lon,
        "min_lat": min_lat,
        "max_lat": max_lat,
        "cell_size_deg": cell_size,
        "nrows": nrows,
        "ncols": ncols,
        "total_points": int(len(lons)),
        "cells_with_data": int(len(counts)),
    }

    # Reorder output columns nicely
    counts_df = counts[["row", "col", "lon_center", "lat_center", "count"]].sort_values(
        ["count"], ascending=False
    )

    return counts_df, grid_meta, grid


def plot_heatmap(grid: np.ndarray, meta: dict, out_png: str):
    """
    Heatmap of counts. Note: this is in lat/lon grid space, not a projected map.
    """
    extent = [meta["min_lon"], meta["max_lon"], meta["min_lat"], meta["max_lat"]]

    plt.figure()
    plt.imshow(
        grid,
        origin="lower",
        extent=extent,
        aspect="auto",
    )
    plt.colorbar(label="Count of railway coordinate points")
    plt.xlabel("Longitude")
    plt.ylabel("Latitude")
    plt.title(f"Railway coordinate density (cell={meta['cell_size_deg']}°)")
    plt.tight_layout()
    plt.savefig(out_png, dpi=200)
    plt.close()


# ---------------------------
# MAIN
# ---------------------------
def main():
    print("Reading:", INPUT_CSV)
    df = pd.read_csv(INPUT_CSV)

    counts_df, meta, grid = build_grid_counts(df, CELL_SIZE_DEG)

    print("Grid summary:", meta)

    print("Writing counts to:", OUT_CSV)
    counts_df.to_csv(OUT_CSV, index=False)

    if MAKE_PLOT:
        print("Writing heatmap to:", OUT_PNG)
        plot_heatmap(grid, meta, OUT_PNG)

    print("Done.")


if __name__ == "__main__":
    main()
```

```python
"""
Equal-size square grid (meters) + counts per cell
Outputs BOTH:
  - CSV
  - XLSX  ✅

Needs:
  pip install pandas numpy pyproj openpyxl
"""

import os
import math
import numpy as np
import pandas as pd
from pyproj import Transformer


# ---------------------------
# USER SETTINGS
# ---------------------------
INPUT_CSV = r"C:\Users\Caleb\Downloads\scraping\railway_lines_1885_new.csv"

CELL_SIZE_M = 10_000   # 10km x 10km equal squares everywhere (change as needed)

OUT_DIR = os.path.dirname(INPUT_CSV)
OUT_CSV  = os.path.join(OUT_DIR, f"grid_counts_equal_{CELL_SIZE_M}m.csv")
OUT_XLSX = os.path.join(OUT_DIR, f"grid_counts_equal_{CELL_SIZE_M}m.xlsx")


# ---------------------------
# HELPERS
# ---------------------------
def parse_coords_string(s: str):
    """coords format: 'lon:lat|lon:lat|...' -> list[(lon, lat)]"""
    if s is None or (isinstance(s, float) and math.isnan(s)):
        return []
    s = str(s).strip()
    if not s:
        return []
    pts = []
    for token in s.split("|"):
        token = token.strip()
        if not token:
            continue
        parts = token.split(":")
        if len(parts) != 2:
            continue
        try:
            lon = float(parts[0])
            lat = float(parts[1])
            pts.append((lon, lat))
        except ValueError:
            continue
    return pts


def main():
    df = pd.read_csv(INPUT_CSV)
    if "coords" not in df.columns:
        raise ValueError("CSV must contain a 'coords' column like 'lon:lat|lon:lat|...'")

    # 1) Extract lon/lat points
    lons, lats = [], []
    for s in df["coords"].values:
        for lon, lat in parse_coords_string(s):
            lons.append(lon)
            lats.append(lat)

    if not lons:
        raise ValueError("No coordinate points found in 'coords' column.")

    lons = np.array(lons, dtype=float)
    lats = np.array(lats, dtype=float)

    # 2) Project to meters (Europe-wide): EPSG:4326 -> EPSG:3035
    # EPSG:3035 is a metric CRS for Europe (good for equal-size grid cells in meters).
    to_m = Transformer.from_crs("EPSG:4326", "EPSG:3035", always_xy=True)
    xs, ys = to_m.transform(lons, lats)
    xs = np.array(xs, dtype=float)
    ys = np.array(ys, dtype=float)

    # 3) Define bounding box in meters
    min_x, max_x = float(xs.min()), float(xs.max())
    min_y, max_y = float(ys.min()), float(ys.max())

    # 4) Assign each point to an equal-size square cell
    col = np.floor((xs - min_x) / CELL_SIZE_M).astype(int)
    row = np.floor((ys - min_y) / CELL_SIZE_M).astype(int)

    ncols = int(col.max()) + 1
    nrows = int(row.max()) + 1

    # 5) Count points per (row, col)
    rc = pd.DataFrame({"row": row, "col": col})
    counts = rc.value_counts().reset_index(name="count")

    # 6) Cell centers (meters)
    counts["x_center_m"] = min_x + (counts["col"] + 0.5) * CELL_SIZE_M
    counts["y_center_m"] = min_y + (counts["row"] + 0.5) * CELL_SIZE_M

    # 7) Also provide lon/lat center for mapping convenience
    to_deg = Transformer.from_crs("EPSG:3035", "EPSG:4326", always_xy=True)
    lon_c, lat_c = to_deg.transform(
        counts["x_center_m"].to_numpy(),
        counts["y_center_m"].to_numpy()
    )
    counts["lon_center"] = lon_c
    counts["lat_center"] = lat_c

    # 8) Stable ID for merging
    counts["cell_id"] = counts["row"].astype(str) + "_" + counts["col"].astype(str)

    out = counts[[
        "cell_id",
        "row", "col",
        "x_center_m", "y_center_m",
        "lon_center", "lat_center",
        "count"
    ]].sort_values("count", ascending=False)

    # Save CSV
    out.to_csv(OUT_CSV, index=False)

    # Save XLSX ✅
    meta = pd.DataFrame([{
        "input_csv": INPUT_CSV,
        "cell_size_m": CELL_SIZE_M,
        "total_points": int(len(xs)),
        "grid_rows": nrows,
        "grid_cols": ncols,
        "cells_with_data": int(len(out)),
        "min_x": min_x, "max_x": max_x,
        "min_y": min_y, "max_y": max_y
    }])

    with pd.ExcelWriter(OUT_XLSX, engine="openpyxl") as writer:
        out.to_excel(writer, sheet_name="grid_counts", index=False)
        meta.to_excel(writer, sheet_name="meta", index=False)

    print("DONE")
    print(f"Points: {len(xs):,}")
    print(f"Cell size: {CELL_SIZE_M:,} m x {CELL_SIZE_M:,} m")
    print(f"Grid: {nrows} rows x {ncols} cols (possible cells: {nrows*ncols:,})")
    print(f"Cells with data: {len(out):,}")
    print("Wrote:")
    print(" -", OUT_CSV)
    print(" -", OUT_XLSX)


if __name__ == "__main__":
    main()
```
